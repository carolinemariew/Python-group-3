```python
import pandas as pd
import chardet

import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import plotly.io as pio
pio.renderers.default = "plotly_mimetype+notebook"
```

```python
path_rail = './Python-group-3/Datasets/rail_pa_speed_page_linear.csv'
with open(path_rail, 'rb') as rawdata:
 result = chardet.detect(rawdata.read(100000))
result
df = pd.read_csv(path_rail, encoding ='Windows-1252')
df
```

```python
path_air = './Python-group-3/Datasets/Air_transport_of_passengers_by_country.csv'
with open(path_air, 'rb') as rawdata:
 result = chardet.detect(rawdata.read(100000))
result
dh = pd.read_csv(path_air, encoding ='Windows-1252')
dh
```

```python
rail = df[df['geo']=="FR"] # the rail data in particular to France is taken from the main dataset
rail
```

```python
air_n = dh[dh['geo']=="FR"]  # similarly, air dataset for particularly France is taken 
air = air_n[air_n['TIME_PERIOD']>2015] # To keep synchronisation data from 2015 is taken 
air['WITHIN_FRANCE'] = air['OBS_VALUE'] * 0.23 # As per the data from European Union, an approximate of 23% of the all air travel is occuring within the nation, that is the domestic flights
air

```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(air['TIME_PERIOD'], air['WITHIN_FRANCE'], label='Air travel', marker='d')  # Both the Air travel and Train travel are compared for France.
ax.plot(rail['TIME_PERIOD'], rail['OBS_VALUE'], label='Railway travel', marker='s') # France is taken specifically due to the presence of High Speed Railway Network in the Country which is competative to the air travel

ax.set_xlabel('Year')
ax.set_ylabel('Total Passengers')
ax.set_title('Comparison for French Air and Railway passengers from 2016 until 2022')
ax.grid(True)
ax.legend()

plt.show()
```
